### Import Packages

```python
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
#Here I load the data and save it as a dataframe using Pandas.
world_bank_df = pd.read_json('world_bank_projects.json')
world_bank_df.head() # This shows the first few rows of the dataframe
```

```python
world_bank_df.countryshortname.value_counts().head(10) #counted using countrys' short names
```

****
### 2.  Top 10 major project themes

```python
with open('world_bank_projects.json') as file: # Context Manager
    data = json.load(file)
    themes_df = json_normalize(data, 'mjtheme_namecode')    
    
major_themes_df = themes_df.name.value_counts().head(10)
major_themes_df
```

****
### 3.  Fill in missing values for theme names

```python
# to have a glance at all the unique code numbers -- code #s are 1-11
themes_df.code.unique() 
```

```python
# create themes dict to pair up the codes with their associated theme names
themes_dict = (themes_df[themes_df.name!=''].drop_duplicates())
themes_dict = dict(zip( themes_dict .code, themes_dict .name))

# fill missing values
themes_df['clean_name']=themes_df.code.apply(lambda item: themes_dict[item])
themes_df.head(30)
```
